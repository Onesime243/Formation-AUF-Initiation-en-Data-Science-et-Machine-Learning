**NumPy**

NumPy (https://numpy.org/), abréviation de *Numerical Python*, est une pierre angulaire du traitement numérique en Python. Elle fournit les structures de données, les algorithmes et la « colle » pour les outils de librairie nécessaires à la plupart des applications scientifiques impliquant des données numériques en Python. NumPy contient, entre autres, les éléments suivants :


* un objet tableau multidimensionnel rapide et efficace, *ndarray* ;
* des fonctions permettant d’effectuer des calculs au niveau des éléments avec des tableaux, ou des opérations mathématiques entre tableaux ;
* des outils de lecture et d’écriture sur disque de jeux de données basés sur des tableaux ;
* des opérations d’algèbre linéaire, de transformée de Fourier et de génération de nombres aléatoires ;
* une interface API C mature pour permettre aux extensions Python et au code natif C ou C++ d’accéder aux structures de données et aux outils de calcul de NumPy

Au-delà des capacités de traitement rapide des tableaux que NumPy ajoute à Python, l’une de ses principales utilisations dans l’analyse des données est de servir de conteneur pour les données à transmettre entre les algorithmes et les librairies. *Pour les données numériques, les tableaux NumPy sont plus efficaces pour le stockage et la manipulation des données que les autres structures de données natives de Python*.

# Structures de données et séquences

Les structures de données de Python sont simples mais puissantes. Maîtriser leur utilisation est essentiel pour devenir un programmeur Python compétent.

**Tuples**

Un tuple (ou n-uplet) est une séquence d’objets Python de longueur fixe et immuable. La façon la plus simple d’en créer un consiste à indiquer une séquence de valeurs séparées par des virgules :


```python
tup = 4, 5, 6
tup
```

Lorsque vous définissez des tuples dans des expressions plus complexes, il est souvent nécessaire de mettre les valeurs entre parenthèses, comme dans cet exemple de création d’un tuple de tuples :

```python
tup_imbriq = (4, 5, 6), (7, 8)
tup_imbriq
```

Vous pouvez convertir n’importe quelle séquence ou itérateur en un tuple en invoquant la fonction *tuple* :

```python
tuple([4, 0, 2])
```

```python
tup = tuple('chaîne')
tup
```

```python
tup[0]
```

```python
tup = tuple(['foo', [1, 2], True])
tup[2] = False
```

Si un objet à l’intérieur d’un tuple est mutable, comme une liste, vous pouvez le modifier sur place :

```python
tup[1].append(3)
tup
```

Vous pouvez concaténer des tuples en utilisant l’opérateur + pour produire des tuples plus longs :

```python
(4, None, 'foo') + (6, 0) + ('bar',)
```

La multiplication d’un tuple par un nombre entier, comme pour les listes, a pour effet de concaténer ensemble le nombre correspondant de copies du tuple :

```python
('foo', 'bar') * 4
```

**Créer un tableau de taille quelconque**

```python
# Créer un tableau de taille m x n rempli de zéros
def creer_tableau(m, n):
    tableau = [[0 for _ in range(n)] for _ in range(m)]
    return tableau

# Exemple d'utilisation
tableau = creer_tableau(3, 4)
print(tableau)
```

## Encodage des Catégories :

LabelEncoder de la bibliothèque scikit-learn est utilisé pour convertir des variables catégorielles en variables numériques, ce qui est souvent nécessaire pour les algorithmes de machine learning qui nécessitent des entrées numériques. Voici comment cela fonctionne :

Fonctionnement de LabelEncoder
Encodage des Catégories :

LabelEncoder attribue un nombre unique à chaque catégorie dans une variable catégorielle.
Par exemple, si une colonne contient les catégories ['Red', 'Blue', 'Green'], LabelEncoder pourrait les convertir en [0, 1, 2].
Ajustement et Transformation :

fit() : Apprend les catégories uniques dans la variable.
transform() : Convertit les catégories en nombres entiers basés sur l'encodage appris.
fit_transform() : Combine les deux étapes ci-dessus en un seul appel.
Décodage :

Vous pouvez également décoder les nombres entiers en leurs catégories d'origine à l'aide de la méthode inverse_transform().

NB. Utiliser One-hot encoding pour les donnees binaires.

#Métriques de base

Lorsqu'on développe un modèle de régression linéaire, il est crucial d'évaluer sa performance afin de déterminer s'il est capable de faire des prédictions précises. Plusieurs métriques sont utilisées à cet effet. Voici quelques-unes des plus importantes.

**Erreur quadratique moyenne (Mean Squared Error, MSE)** : Calcule la moyenne des carrés des différences entre les valeurs prédites et les valeurs réelles. Une valeur MSE basse indique un bon ajustement du modèle.

**Racine carrée de l'erreur quadratique moyenne (Root Mean Squared Error, RMSE)** : Représente l'écart type des résidus. Elle est souvent préférée à la MSE car elle est exprimée dans les mêmes unités que la variable cible.

**Erreur absolue moyenne (Mean Absolute Error, MAE)** : Calcule la moyenne des valeurs absolues des différences entre les valeurs prédites et les valeurs réelles. Elle est moins sensible aux valeurs aberrantes que la MSE.

**Coefficient de détermination (R²)**
R² : Mesure la proportion de la variance de la variable dépendante expliquée par le modèle. Une valeur de R² proche de 1 indique que le modèle explique bien la variabilité des données.

# Comparaison des modèles  

```python
# import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
```

```python
df = pd.read_csv("C:/Users/Onesime/Documents/DATA SCIENCE AND MACHINE LEARNING TRAINING FILES_AUF BY Onesime_mb/Workshop Data Science_ML by Onesime Mbulayi Files AUF_2025/IRIS.csv")
df.head()
```

```python
X = df.drop(['species'], axis=1)
Y = df['species']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=7)
```

```python
print("X_train.shape:", X_train.shape)
print("X_test.shape:", X_test.shape)
print("Y_train.shape:", X_train.shape)
print("Y_test.shape:", Y_test.shape)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# modèles
models = []

# modèles linéaires
models.append(('LR', LogisticRegression(solver='liblinear', multi_class="auto")))
models.append(('LDA', LinearDiscriminantAnalysis()))

# modèles non linéaires
models.append(('CART', DecisionTreeClassifier()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('GNB', GaussianNB()))
models.append(('SVC', SVC(gamma="auto")))

# Evaluer chaque modèle à tour de rôle
print("Model Accuracy:")
names = []
accuracy = []
for name, model in models:
    # Validation croisée en 10 étapes pour évaluer le modèle
    kfold = KFold(n_splits=10)
    cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
    
    # Afficher les résultats de la validation croisée du modèle actuel
    names.append(name)
    accuracy.append(cv_results)
    msg = "%s: accuracy=%f std=(%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```
